```python
import pandas as pd
from surprise import SVD
from surprise.model_selection import cross_validate, train_test_split
from surprise import accuracy
```

```python
grid = pd.read_csv('user_game.csv')
```

```python
max_value_username = pd.DataFrame(grid.groupby('user_name')['max'].max().reset_index())
```

```python
max_val_dict = max_value_username.groupby('user_name')['max'].apply(list).to_dict()
```

```python
grid['max_game'] = grid['user_name'].map(max_val_dict)
grid.info()
```

```python
grid['max_game_int'] = grid.max_game.str[0].astype(int)
grid = grid.drop('max_game', axis = 1)
grid.info()
```

```python
grid['score'] = grid['max']/grid['max_game_int']
grid.sample(3)
```

```python
from sklearn.preprocessing import minmax_scale
grid['scaled_score'] = minmax_scale(grid['score'], feature_range=(1, 5))
```

```python
grid.info()
```

```python
grid = grid.dropna()
```

```python
#grid.to_csv('surprise_grid.csv', index = False)
```

```python
from __future__ import (absolute_import, division, print_function,
                        unicode_literals)
import surprise
from surprise import SVDpp
import random
from surprise import SVD
from surprise import Dataset
from surprise import accuracy
from surprise.model_selection import GridSearchCV
```

```python
reader = surprise.Reader(rating_scale = (1, 5))
```

```python
data = Dataset.load_from_df(grid[['user_name', 'game_name', 'scaled_score']], reader)
```

```python
alg = SVDpp(verbose=True)
```

```python
cross_validate(alg, data, measures=['RMSE', 'MAE', 'FCP'], cv=5, n_jobs=-1, verbose=True)
```

```python
trainset, testset = train_test_split(data, test_size=.25)

# Train the algorithm on the trainset, and predict ratings for the testset
alg.fit(trainset)
predictions = alg.test(testset)
```

```python
surprise.accuracy.fcp(predictions, verbose=True)
```

```python
from collections import defaultdict
```

```python
def get_top_n(predictions, n=10):
    top_n = defaultdict(list)
    for uid, iid, true_r, est, _ in predictions:
        top_n[uid].append((iid, est))

    # Then sort the predictions for each user and retrieve the k highest ones.
    for uid, user_ratings in top_n.items():
        user_ratings.sort(key=lambda x: x[1], reverse=True)
        top_n[uid] = user_ratings[:n]

    return top_n
```
