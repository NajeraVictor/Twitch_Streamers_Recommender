# Scraping Box Office Mojo with Selenium

```python
chromedriver = "/Applications/chromedriver" # path to the chromedriver executable
os.environ["webdriver.chrome.driver"] = chromedriver 

driver = webdriver.Chrome(chromedriver)

matrix_url = "https://www.twitch.tv/directory"
driver.get(matrix_url) #opens the matrix in the dummy google chrome
```

```python
# 'contains' will find a match on the text, in this case return b tag
game_list = []
num = 2
while num < 1000:
    gross_selector = f'//*[@id="root"]/div/div[2]/div/main/div/div[3]/div/div/div/div[3]/div/div/div[1]/div[{num}]/div/div/div/div[1]/div/a/div[2]/div/div/h3'
    print(driver.find_element_by_xpath(gross_selector).text)
    game_list.append(driver.find_element_by_xpath(gross_selector).text)    
    time.sleep(.5+2*random.random())
    num+=1

```

```python
from selenium.common.exceptions import *
# 'contains' will find a match on the text, in this case return b tag
game_list_cats = []

#this loop ends at retro games because retro games has no tag
num = 2
while num < 1000:
    try:
        gross_selector = f'//*[@id="root"]/div/div[2]/div/main/div/div[3]/div/div/div/div[3]/div/div/div[1]/div[{num}]/div/div/div/div[2]'
        print(driver.find_element_by_xpath(gross_selector).text.split('\n'))
        game_list_cats.append(driver.find_element_by_xpath(gross_selector).text.split('\n'))
        time.sleep(.5+2*random.random())
        num+=1
    except NoSuchElementException:
        print(['NONE'])
        game_list_cats.append(['NONE'])
        time.sleep(.5+2*random.random())        
        num+=1
```

```python
games = [item for item in zip(game_list,game_list_cats)]
```
