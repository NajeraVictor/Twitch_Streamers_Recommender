# Use MAX( viewers ) in general_view table to filter out duplicate rows (from genre)

```python

query = '''SELECT user_id, user_name, count(distinct(game_id)) as num_games, MAX(viewer_count) as max_viewers
FROM stream_data 
GROUP BY user_id, user_name
HAVING count(distinct(game_id))>1
ORDER BY num_games DESC, max_viewers DESC;'''
pd_sql.read_sql(query, connection)
```

# Find number of viewers in top 100 streamers per hour by game

```python
# Find number of viewers in top 100 streamers per hour by game
query = '''
SELECT time_logged, game_name, sum(max)
FROM (SELECT time_logged, stream_id, game_name, (MAX(viewer_count)) 
    FROM general_view
    GROUP BY time_logged, game_name, stream_id
    ORDER BY time_logged DESC , game_name
    ) as bob
GROUP BY time_logged, game_name
ORDER BY game_name, time_logged DESC, sum DESC
'''
df=pd_sql.read_sql(query, connection)
```

# Find max viewers a streamer has ever had on a stream and in what game (in our data set)

```python
query = '''
SELECT user_name, game_name, max(viewer_count)
FROM general_view
GROUP BY user_name, game_name
ORDER BY user_name DESC, max DESC
LIMIT 10000
'''
df=pd_sql.read_sql(query, connection)
```
