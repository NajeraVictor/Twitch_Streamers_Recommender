```python
genres = genres.category.apply(pd.Series) \
    .merge(genres, right_index = True, left_index = True) \
    .drop(["category"], axis = 1) \
    .melt(id_vars = ['game'], value_name = "category") \
    .drop("variable", axis = 1) \
    .dropna()
```
